```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
from time import sleep
from random import randint
import re

```

```python
# get all the page suffix
page_lst = []
main_page = requests.get('https://www.beeradvocate.com/beer/top-rated/')

soup = BeautifulSoup(main_page.text, 'html.parser')

pattern = "/beer/profile/(.+?)/(.+?)/"

pages_link = soup.findAll('a',attrs={'href': re.compile(pattern)})

pages_data = []

for l in pages_link:
    l = str(l)
    m = re.search(pattern,l)
    pages_data.append(m.group(1)+'/'+m.group(2))
    print(l)
```

```python
# Scraping product_name, product_review, and user_rating
product_name = []
data = pd.DataFrame()

for p in pages_data:

    review = []
    rating = []

    page = requests.get('https://www.beeradvocate.com/beer/profile/'+p+'/')
    soup = BeautifulSoup(page.text,'html.parser')

    # product_names
    name = soup.find('div',class_ ="titleBar")
    search = re.search('<h1>(.+?)<br/>',str(name))
    name = search.group(1)
    product_name.append(name)

    # progress checker
    print(name,len(product_name))


    comments = soup.findAll('div', attrs = {'id':'rating_fullview_content_2'})

    sleep(randint(2,8))

    for c in comments:

        # comments
        com = c.find('div')
        if com != None:
            cc = re.search('\">(.*?)</div>',str(com),flags = re.S)
            cc1 = str(cc.group(1)).replace('<br>','').replace('</br>','')
            review.append(cc1)
        else:
            review.append('')

        # ratings
        score = c.find('span', attrs = {'class':'BAscore_norm'})

        if score == None:
            score = c.find('span', attrs = {'class':'muted'})
            s = re.search('Rated:(.+?)by',str(score))
            s = str(s.group(1)).replace('<b>','').replace('</b>','')
            rating.append(float(s))
        else:
            rating.append(float(score.text))

    names = [name for i in range(len(rating))]

    d1 = pd.DataFrame({'product_name':names,'product_review':review,'user_rating':rating})
    data = pd.concat([data,d1])


data.reset_index(drop = True, inplace = True)
```

```python
from collections import Counter
import nltk
from nltk.corpus import stopwords
import string

# Download the NLTK stop words
nltk.download('stopwords')

# Combine all product reviews into a single text and convert to lowercase
all_reviews_text = ' '.join(beer_data['product_review'].astype(str)).lower()

# Remove punctuation from the text
translator = str.maketrans('', '', string.punctuation)
all_reviews_text = all_reviews_text.translate(translator)

# Remove <br/> tag from the text
all_reviews_text = all_reviews_text.replace('<br/>', ' ')

# Tokenize the text (split it into words)
words = all_reviews_text.split()

# Create a word frequency dictionary
word_freq = Counter(words)

# Define a list of common English stop words
stop_words = set(stopwords.words('english'))

# Define a custom list of words to exclude and convert to lowercase
custom_excluded_words = ['nan', 'the', 'this', 'and', 'with', 'for', 'that', 'from', 'but', 'not',
                        'beer', 'taste', 'like', 'one', 'really', 'much', 'feel', 'overall',
                        'quite', "it's", 'also', 'finger', 'would', 'almost', 'back', 'first', 'br', 'get']

custom_excluded_words = [word.lower() for word in custom_excluded_words]

# Remove common stop words, custom excluded words, and words with length less than 2
filtered_words = [word for word in words if word not in stop_words and word not in custom_excluded_words and len(word) > 1]

# Create a word frequency dictionary for filtered words
filtered_word_freq = Counter(filtered_words)

# Find the most common attributes by setting a threshold for word frequency
min_word_freq = 100  # Adjust this threshold as needed
most_common_attributes = [(word, freq) for word, freq in filtered_word_freq.items() if freq >= min_word_freq]

# Sort the most_common_attributes list by frequency in descending order
most_common_attributes.sort(key=lambda x: x[1], reverse=True)

# Display the most common attributes and their frequencies
print(f"Most Common Attributes (with frequency >= {min_word_freq}):")
for word, freq in most_common_attributes:
    print(f"{word}: {freq} times")
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Read the product attributes and reviews
attributes_df = pd.read_csv('attributes_file.csv')
attributes_df['attributes'] = attributes_df[['attribute1', 'attribute2', 'attribute3']].agg(' '.join, axis=1)
attribute1 = attributes_df['attribute1'].iloc[0].lower()
attribute2 = attributes_df['attribute2'].iloc[0].lower()
attribute3 = attributes_df['attribute3'].iloc[0].lower()

beer_data = beer_data.fillna('')
beer_data['product_review'] = beer_data['product_review'].str.lower()

# List to hold results
results = []

for _, row in beer_data.iterrows():
    product_name = row['product_name']
    product_review = row['product_review']

    # Combine attributes and product review for comparison
    documents = [attribute1, attribute2, attribute3, product_review]
    count_vectorizer = CountVectorizer(stop_words='english')
    sparse_matrix = count_vectorizer.fit_transform(documents)

    # Taking the average of the similarity scores with all three attributes
    avg_similarity_score = sum(cosine_similarity(sparse_matrix[-1], sparse_matrix[:-1]).flatten()) / 3

    results.append({
        'product_name': product_name,
        'product_review': product_review,
        'similarity_score': avg_similarity_score
    })

# Convert results to DataFrame
results_df = pd.DataFrame(results)

# Save results to an output file
output_file = 'output.csv'
results_df.to_csv(output_file, index=False)
```

### Task H

The top 4 attributes of beer we got are: **'head', 'dark', 'chocolate', ' coffee'**.

Utilizing these four attributes, we calculate the lift values between top 10 beers with the most reviews.

```python
beer_data = beer_data.dropna()

# Calculate Lift Values
def lift_calculation(data, beer, attr):
    total = len(data)
    attributes_df = data[data['product_review'].str.contains(attr, na=False, case=False)]
    beer_df = data[data['product_name'] == beer]
    share = data[(data['product_review'].str.contains(attr, na=False, case=False))&(data['product_name'] == beer)]

    p_attr = len(attributes_df)/total
    p_beer = len(beer_df)/total
    p_two = len(share)/total

    return p_two / (p_attr*p_beer)

# Get top 10 beers
top_10_beer = beer_data['product_name'].value_counts().sort_values(ascending=False).head(10).index.to_list()
# Top 4 Attributes
top_4_attr = ['head','dark','chocolate','coffee']
# Create DataFrame to store lift values
lift = pd.DataFrame(index=top_10_beer, columns=top_4_attr)
# Calculate Lift
for beer in top_10_beer:
    for attr in top_4_attr:
        lift.loc[beer, attr] = lift_calculation(beer_data, beer, attr)
lift
```

We decided to use Euclidean Distance as our meausring method to find the beer with highest similarity with the target beer. We chose Euclidean Distance since it works the best when we have lift values for multiple pairs of items and want to compare their relationships.

```python
import numpy as np
# Get the chosen beer
target_beer = top_10_beer[0]
# Create list to store siilarity score between chosen beer and other beers
similarity = []

# Get similarities
for beer in top_10_beer[1:]:
    s = np.linalg.norm(np.array([lift.loc[target_beer].values]) - np.array([lift.loc[beer].values])) # get Euclidean Distance
    similarity.append(s)

# Display result
result = {'target_product':[target_beer for i in range(9)],'product_name':top_10_beer[1:], 'Euclidean_Distance':similarity}
result = pd.DataFrame(data=result)
result.sort_values(['Euclidean_Distance'])
```
